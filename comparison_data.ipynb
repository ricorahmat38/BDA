```python
# Hasil dari pengolahan data manual (misalnya dari Notebook 1)
data_ingestion = {
    'accuracy': 0.20189274447949526,
    'precision': 0.19816837371723514,
    'recall': 0.20189274447949526,
    'f1_score': 0.19709166526757918,
    'conf_matrix': [[9, 17, 24], [17, 7, 98], [5, 92, 48]]
}

# Hasil dari pengolahan menggunakan Random Forest (misalnya dari Notebook 2)
data_randomforest = {
    'accuracy': 1.0,
    'precision': 1.0,
    'recall': 1.0,
    'f1_score': 1.0,
    'conf_matrix': [[50, 0, 0], [0, 122, 0], [0, 0, 145]]
}
```

```python
import pandas as pd

# Membuat DataFrame untuk perbandingan metrik
metrics_comparison = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Manual': [data_ingestion['accuracy'], data_ingestion['precision'], data_ingestion['recall'], data_ingestion['f1_score']],
    'Random Forest': [data_randomforest['accuracy'], data_randomforest['precision'], data_randomforest['recall'], data_randomforest['f1_score']]
})

# Menampilkan perbandingan
metrics_comparison
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Matriks Kebingunguan Manual
data_ingestion = data_ingestion['conf_matrix']

# Matriks Kebingunguan Random Forest
data_randomforest = data_randomforest['conf_matrix']

# Membuat plot perbandingan confusion matrix
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.heatmap(data_ingestion, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['Prediksi Negatif', 'Prediksi Positif', 'Prediksi Netral'],
            yticklabels=['Aktual Negatif', 'Aktual Positif', 'Aktual Netral'])
axes[0].set_title('Confusion Matrix Manual')

sns.heatmap(data_randomforest, annot=True, fmt='d', cmap='Blues', ax=axes[1],
            xticklabels=['Prediksi Negatif', 'Prediksi Positif', 'Prediksi Netral'],
            yticklabels=['Aktual Negatif', 'Aktual Positif', 'Aktual Netral'])
axes[1].set_title('Confusion Matrix Random Forest')

plt.tight_layout()
plt.show()

```
