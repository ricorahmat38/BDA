```python
import pandas as pd
from collections import Counter

file_path = 'rico/teks/sipsn/Data_Komposisi_sumber_sampah_SIPSN_Surabaya.xlsx'
data = pd.read_excel(file_path, engine='openpyxl')

combined_data = data.apply(lambda x: ' '.join(x.astype(str)), axis=1)

def word_frequencies(doc):
    words = doc.split()
    return Counter(words)

document_frequencies = combined_data.apply(word_frequencies)

def average_word_count(doc_freq):
    total_words = sum(doc_freq.values()) 
    num_words = len(doc_freq)  
    return total_words / num_words if num_words > 0 else 0  

average_counts = document_frequencies.apply(average_word_count)

mean_value = average_counts.mean()
min_value = average_counts.min()
max_value = average_counts.max()

print(f"Rata-rata kemunculan kata per dokumen: {mean_value}")
print(f"Nilai minimum kemunculan kata per dokumen: {min_value}")
print(f"Nilai maksimum kemunculan kata per dokumen: {max_value}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Memuat data dari file Excel
file_path = 'rico/teks/sipsn/Data_Komposisi_sumber_sampah_SIPSN_Surabaya.xlsx'
data = pd.read_excel(file_path, engine='openpyxl')

# Menggabungkan semua kolom menjadi satu kolom teks per baris
# Pastikan hanya menggabungkan kolom yang berisi teks atau string
# Menggunakan .astype(str) untuk memastikan semua data diperlakukan sebagai string
data_combined = data.astype(str).agg(' '.join, axis=1)

# Inisialisasi TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Mengubah teks gabungan menjadi vektor TF-IDF
tfidf_matrix = vectorizer.fit_transform(data_combined)

# Menampilkan vektor TF-IDF sebagai DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Menampilkan DataFrame TF-IDF
print(tfidf_df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import OneHotEncoder

file_path = 'rico/teks/sipsn/Data_Komposisi_sumber_sampah_SIPSN_Surabaya.xlsx'
data = pd.read_excel(file_path, engine='openpyxl')

X = data.drop(columns=['Tahun'])
y = data['Tahun']

X_encoded = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=1)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)
conf_matrix = confusion_matrix(y_test, y_pred)


print(f"Akurasi: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
print("Confusion Matrix:")
print(conf_matrix)
```
